```python
import seaborn as sns
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('housing.csv')
df.head()
```

```python
house_df = df[df['Type'] == 'house']
apartment_df = df[df['Type'] == 'apartment']
unit_df = df[df['Type'] == 'unit']
townhouse_df = df[df['Type'] == 'townhouse']

plt.scatter(house_df['Landsize'], house_df['Price'], color="Green")
plt.scatter(apartment_df['Landsize'], apartment_df['Price'], color="Blue")
plt.scatter(unit_df['Landsize'], unit_df['Price'], color="Red")
plt.scatter(townhouse_df['Landsize'], townhouse_df['Price'], color="Yellow")

plt.show()
```

```python
year_df = df.groupby('YearBuilt', as_index=False).agg({'Price': 'mean'})
year_df
```

```python
plt.figure(figsize=(16, 8), dpi=1000)
sns.lineplot(x='YearBuilt', y='Price', data=year_df)
plt.show()
```

```python
region_df = df.groupby('Regionname', as_index=False).agg({'Price': 'mean'})
region_df
```

```python
region_Eastern_Metropolitan_df = df[df['Regionname'] == 'Eastern Metropolitan']
region_Western_Victoria_df = df[df['Regionname'] == 'Western Victoria']
region_Southern_Metropolitan_df = df[df['Regionname'] == 'Southern Metropolitan']
region_Northen_Metropolitan_df = df[df['Regionname'] == 'Northen Metropolitan']

region_data = [region_Eastern_Metropolitan_df['Price'], region_Western_Victoria_df['Price'], region_Southern_Metropolitan_df['Price'], region_Northen_Metropolitan_df['Price']]

plt.boxplot(region_data)
plt.show()
```

```python
sns.histplot(x='Type',y='Price', data=df)
plt.show()
```

```python
sns.boxplot(x='YearBuilt', y='Regionname', data=df, orient='h')
plt.show()
```
